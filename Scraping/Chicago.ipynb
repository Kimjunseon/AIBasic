```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/'

# 해당 페이가 request를 거부 할 때..
html = requests.get(url, headers={"User-Agent":"Mozilla/5.0"}).text
soup = BeautifulSoup(html, 'lxml')
```

```python
print(soup)
```

# 목표 1: 랭크, 메인메뉴, 카페이름, 사이트주소, 가격, 위치 값 가져오기
# 목표 2: 구글맵 표시

```python
# 랭크 뽑기
value_ranks = soup.select('div.sammy div.sammyRank')
```

```python
# 메인 메뉴 뽑기
value_menus = soup.select('div.sammy b')
```

```python
# 카페이름 뽑기 ([0]은 예시)
value_cafe_names = soup.select('div.sammyListing')[0].text.split('\n')
```

```python
# 사이트 주소 뽑기 ([0]은 예시)
value_site_url = soup.select('div.sammy a')[0].get('href')
```

```python
rank = []
menu = []
cafe_name = []
site_url= []
tmp = 0

for value_rank in value_ranks:
    rank.append(value_ranks[tmp].text)
    menu.append(value_menus[tmp].text)
    value_cafe_names = soup.select('div.sammyListing')[tmp].text.split('\n')
    cafe_name.append(value_cafe_names[1])
    site_url.append(soup.select('div.sammy a')[tmp].get('href'))
    tmp += 1
```

```python
base_url = 'https://www.chicagomag.com'
```

```python
# 사이트 조인 하기

from urllib.parse import urljoin
from tqdm.notebook import tqdm

full_site_url= []
tmp = 0

for value_rank in tqdm(value_ranks): 
    full_site_url.append(urljoin(base_url, soup.select('div.sammy a')[tmp].get('href')))
    tmp += 1
```

```python
import pandas as pd
data = {'Rank':rank, 'Cafe':cafe_name, 'Menu':menu, 'URL':full_site_url}
df = pd.DataFrame(data)
df.head()
```

```python
html = requests.get(df['URL'][0], 
                headers={"User-Agent":"Mozilla/5.0"}).text
el = BeautifulSoup(html, 'lxml')
```

```python
import time
price = []
address = []

for n in tqdm(df.index):
    html = requests.get(df['URL'][n], 
                headers={"User-Agent":"Mozilla/5.0"})
    time.sleep(1)
    el = BeautifulSoup(html.text, 'lxml')
    val_list = el.select('p.addy')\
                [0].text.split(',')[0].split()
    price.append(val_list[0][:-1])
    address.append(' '.join(val_list[1:]))
```

```python
print(price)
```

```python
print(address)
```

```python
df['Price'] = price
df['Address'] = address
df.head()
```

```python
import folium
import googlemaps
import numpy as np
```

```python
gmaps_key = "A2BC3DE3FG3HI3JK3LM3NM3NL3K"
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
lat = []
lng = []

for n in tqdm(df.index):
    if df['Address'][n] != 'Multiple locations':
        target_name = df['Address'][n] + ', ' + 'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        time.sleep(0.5)
        location = gmaps_output[0].get('geometry')
        lat.append(location['location']['lat'])
        lng.append(location['location']['lng'])
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```
