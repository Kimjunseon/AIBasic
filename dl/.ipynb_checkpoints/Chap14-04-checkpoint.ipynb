```python
import pandas as pd
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import os
```

```python
df = pd.read_csv('data/wine.csv', header=None)
X = df.iloc[:, :12]
y = df.iloc[:, 12]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
model = models.Sequential()
model.add(layers.Dense(30, input_dim=12, activation='relu'))
model.add(layers.Dense(12, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# patience:20 번 연속적으로 loss 값이 줄어들지않는다면 학습을 강제 중지한다.
early_stopping = EarlyStopping(monitor='val_loss', patience=20)
model_path = './model/best_model.hdf5'

# 상태가 좋은 값이 들어올 때 마다 오버라이드한다.
checkpoint = ModelCheckpoint(filepath=model_path, monitor='val_loss', verbose=1,
                            save_best_only=True)
```

```python
history = model.fit(X_train, y_train, epochs=2000, batch_size=500, verbose=1, validation_split=0.25,
                   callbacks=[early_stopping, checkpoint])
```

```python
score = model.evaluate(X_test, y_test)
print('Test Accuracy:', score[1])
```
